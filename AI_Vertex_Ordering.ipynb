First extract the data from the .txt file and store it in data[ ]. Here i have also set the value of n = the number of vertices. I have used data3.txt so n=19.

```python
import re

# for convenience just change values here when changing the dataset
n = 19
file_path = 'data3.txt'

def read_data(file_path):

    data = [] # this is where i will store the data after extracting it from the file

    with open(file_path, 'r') as file:

        for line in file:

            parts = line.strip().split(',') # we get the vertex, its parent set, and cost in three parts

            vertex = int( parts[0] ) # the vertex
            parent_set = set( map( int, parts[1].strip('{}').split() ) ) if parts[1] != '{}' else set() # the parent set
            cost = float( re.sub( r'[^\d.]', '', parts[2] ) ) # using regex to get the cost

            data.append( (vertex, parent_set, cost) ) # put all that in our data array

    return data

data = read_data(file_path)
print(data)
```

1. Breadth First Search

```python
from collections import deque

def bfs(data, start_vertex, display=0):
    # jsut the usual initialization for bfs
    total=0
    visited = set()
    queue = deque([(start_vertex, [], 0)])

    # breadth first search
    while queue:
        vertex, path, cost = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            if display: # the searching details are only printed if the display flag is set so that there is no unnecessary long output
              print("Vertex:", vertex, "Path:", path, "Cost:", cost)
            total += cost

            for v, parent_set, edge_cost in data: # finding vertex neighbours
                if v == vertex: # so that it only checks neighbours of the vertex we popped from the queue
                    for neighbor in parent_set:
                        if neighbor not in visited:
                            queue.append((neighbor, path + [vertex], cost + edge_cost))
    #print("total cost = ", total)
    return total

# trying the algorithm with each node as source, and finding which one results in least cost
min = bfs(data, 1)
idx = 1
for i in range(1, n+1):
  cost = bfs(data, i)
  if cost < min:
    min = cost
    idx = i

print("Total cost: ", bfs(data, idx, 1))
```

2. Depth First Search

```python
def dfs(data, start_vertex, display=0):
  #initialization
    total = 0
    visited = set()

    def dfs_helper(vertex, path, cost):
        nonlocal total #nonlocal so that it uses the same variable total from outer scope
        if vertex not in visited:
            visited.add(vertex)
            if display:
              print("Vertex:", vertex, "Path:", path, "Cost:", cost)
            total += cost

            for v, parent_set, edge_cost in data:
                if v == vertex:
                    for neighbor in parent_set:
                        if neighbor not in visited:
                            dfs_helper(neighbor, path + [vertex], cost + edge_cost) # recursive dfs call

    dfs_helper(start_vertex, [], 0)
    #print("Total cost for visited vertices:", total)
    return total

# trying the algorithm with each node as source, and finding which one results in least cost
min = dfs(data, 1)
idx = 1
for i in range(1, n+1):
  cost = dfs(data, i)
  if cost < min:
    min = cost
    idx = i

print("Total cost: ", dfs(data, idx, 1)) # the 3rd arg is 0 by default, passing 1 tells the function to display the paths as it keeps adding
```

3. Uniform Cost Search

```python
from queue import PriorityQueue

def ucs(data, start_vertex, display=0):
  # initialization
    total = 0
    visited = set()
    pq = PriorityQueue() # to order vertices by cost
    pq.put((0, start_vertex, []))

    while not pq.empty():
        cost, vertex, path = pq.get()
        if vertex not in visited:
            visited.add(vertex)
            print("Vertex:", vertex, "Path:", path, "Cost:", cost)
            total += cost

            for v, parent_set, edge_cost in data:
                if v == vertex:
                    for neighbor in parent_set:
                        if neighbor not in visited:
                            pq.put((cost + edge_cost, neighbor, path + [vertex])) # add to queue
    #print("Total cost for visited vertices:", total)
    return total

# trying the algorithm with each node as source, and finding which one results in least cost
min = ucs(data, 1)
idx = 1
for i in range(1, n+1):
  cost = ucs(data, i)
  if cost < min:
    min = cost
    idx = i

print("Total cost: ", ucs(data, idx, 1)) # the 3rd arg is 0 by default, passing 1 tells the function to display the paths as it goes along
```
